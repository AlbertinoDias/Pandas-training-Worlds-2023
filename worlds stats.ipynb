<font size=10>Functions to control the csv<font>

```python
def check_str(is_string):
    if isinstance(is_string, str):
        return True
    else: return False
```

```python
def check_bool(is_bool):
    if isinstance(is_bool, bool):
        return True
    else: return False
```

```python
def check_numeric(is_numeric):
    if isinstance(is_numeric, int):
        return True
    else: return False
```

```python
def check_date(is_date):
    if isinstance(is_date, dt.date):
        return True
    else: return False
```

```python
def check_time(is_time):
    if isinstance(is_time, dt.time):
        return True
    else: return False
```

```python
def check_side(side):
    if check_str(side):
        if side == "R" or side == "B":
            return True
    else: return False
```

```python
def check_positive(number):
    if check_numeric(number):
        if number < 0:
            return False
        else: return True
    else: return False
```

```python
def check_order(number):
    if check_numeric(number):
        if number < 1 or number > 5:
            return False
        else: return True
    else: return False
```

```python
def add_champion(champions, champion,picked,side,order,team,enemy,game,position,player,fb,kills,deaths,assists,date,won,time,penta,stage):
    if check_str(champion):
        if check_bool(picked):
            if check_side(side):
                if check_order(order):
                    if check_str(team):
                        if check_str(enemy):
                            if check_order(game):
                                if check_str(position):
                                    if check_str(player):
                                        if check_bool(fb):
                                            if check_positive(kills):
                                                if check_positive(deaths):
                                                    if check_positive(assists):
                                                        if check_date(date):
                                                            if check_bool(won):
                                                                if check_time(time):
                                                                    if check_positive(penta):
                                                                        if check_str(stage):
                                                                            champion = champion.upper()
                                                                            side = side.upper()
                                                                            team = team.upper()
                                                                            enemy = enemy.upper()
                                                                            position = position.upper()
                                                                            player = player.upper()
                                                                            stage = stage.upper()
                                                                            list_to_add = pd.DataFrame({"champion": [champion], "picked":[picked], "side":[side],
                                                                                  "order":[order], "team":[team], "enemy":[enemy],
                                                                                  "game":[game], "position":[position], "player":[player],
                                                                                  "fb":[fb], "kills":[kills], "deaths":[deaths],
                                                                                  "assists":[assists], "date":[date],
                                                                                    "won":[won], "time":[time], "penta":[penta], "stage":[stage]})
                                                                            champions = pd.concat([champions, list_to_add], ignore_index = True)
                                                                            return champions
                                                                            print("working")
    else: print("not working")
```

```python
def add_drake(drakes, name,team,enemy,game,order,time,date,won,stage):
    if check_str(name):
        if check_str(team):
            if check_str(enemy):
                if check_order(game):
                    if check_numeric(order):
                        if check_time(time):
                            if check_date(date):
                                if check_bool(won):
                                    if check_str(stage):
                                        name = name.upper()
                                        team = team.upper()
                                        enemy = enemy.upper()
                                        stage = stage.upper()
                                        list_to_add = pd.DataFrame({"type": [name], "team":[team], "enemy":[enemy],
                                                                          "game":[game], "order":[order], "time":[time],
                                                                          "date":[date], "won": [won], "stage": [stage]})
                                        drakes = pd.concat([drakes, list_to_add], ignore_index = True)
                                        return drakes
                                        print("working")
    else: print("not working")
```

<font size=5>champions time modifier<font>

```python
def modify_champions_date_everything(function):
    def wrapper():
        global champions
        global less_than
        champions_copy = champions.copy()
        function(champions_copy)
    return wrapper 

def modify_champions_date_before(function):
    def wrapper():
        global champions
        global less_than
        champions_copy = champions.copy()
        champions_copy["date"] = pd.to_datetime(champions_copy["date"], format="%Y-%m-%d")
        champions_copy = champions_copy.loc[(champions_copy['date'] <= less_than)]
        function(champions_copy)
    return wrapper    


def modify_champions_exact_date(function):
    def wrapper():
        global champions
        global exact_date
        champions_copy = champions.copy()
        champions_copy['date'] = pd.to_datetime(champions_copy['date'], format='%Y-%m-%d')
        champions_copy = champions_copy.loc[champions_copy['date'] == exact_date]
        function(champions_copy)
    return wrapper



def modify_champions_date_between(function):
    def wrapper():
        global champions
        global between_time
        
        champions_copy = champions.copy()
        champions_copy["date"] = pd.to_datetime(champions_copy["date"], format="%Y-%m-%d")
        champions_copy = champions_copy.loc[(champions_copy['date'] <= between_time[1]) & (champions_copy['date'] >= between_time[0])]

        function(champions_copy)
    return wrapper
```

```python
@modify_champions_date_everything
#@modify_champions_date_before
#@modify_champions_date_between
#@modify_champions_exact_date
def timed_champions_highest_winrate(champions_copy):
    champions_picked = champions_copy.where(champions_copy["picked"] == True).dropna()
    champions_picked_5_p = champions_picked.value_counts(subset = ["champion"])
    champions_picked_5_p = champions_picked_5_p.reset_index()
    champions_picked_5_p = champions_picked_5_p.rename(columns={champions_picked_5_p.columns[1]: 'pick'})
    champions_picked_5_p = champions_picked_5_p.where(champions_picked_5_p["pick"] > 4).dropna()
    
    number_of_times_champions_picked_won = champions_picked.where(champions_copy["won"] == True).value_counts(subset = ["champion"])
    number_of_times_champions_picked_won = number_of_times_champions_picked_won.reset_index()
    number_of_times_champions_picked_won = number_of_times_champions_picked_won.rename(columns={number_of_times_champions_picked_won.columns[1]: 'won'}) 

    champion_winrate_5_P = pd.merge(champions_picked_5_p, number_of_times_champions_picked_won, on='champion')
    champion_winrate_5_P["percentage_of_wins"] = champion_winrate_5_P["won"] * 100 / champion_winrate_5_P["pick"]
    champion_winrate_5_P = champion_winrate_5_P.sort_values("percentage_of_wins", ascending = False ).head(10)
    
    display(champion_winrate_5_P)

timed_champions_highest_winrate()
```

```python
#this code does not verify if deaths are always 0
@modify_champions_date_everything
#@modify_champions_date_before
#@modify_champions_date_between
#@modify_champions_exact_date
def timed_hightest_kda_per_player(champions_copy):
    k_d_a_by_player = champions_copy[["player","kills", "deaths","assists"]].where(champions_copy["picked"] == True).dropna()
    k_d_a_by_player = k_d_a_by_player.groupby("player").sum()
    k_d_a_by_player = k_d_a_by_player.where(k_d_a_by_player["deaths"] > 0)

    kda_by_player = k_d_a_by_player

    kda_by_player["kda"] = (kda_by_player["kills"] + kda_by_player["assists"]) / kda_by_player["deaths"]
    kda_by_player["kda"] = kda_by_player["kda"].round(1)
    kda_by_player = kda_by_player.sort_values("kda", ascending = False).head(10)
    display(kda_by_player)
timed_hightest_kda_per_player()
```

```python
@modify_champions_date_everything
#@modify_champions_date_before
#@modify_champions_date_between
#@modify_champions_exact_date
def timed_blue_v_red_winrate(champions_copy):
    side_winrate = champions_copy[["team","enemy","side","won","date","game"]].dropna().drop_duplicates()

    r_side_winrate = side_winrate.where(side_winrate["side"] == "R").dropna()
    r_side_winrate = r_side_winrate.where(r_side_winrate["won"] == True).dropna()
    r_side_winrate = r_side_winrate.value_counts(subset = ["side"]).head()


    b_side_winrate = side_winrate.where(side_winrate["side"] == "B").dropna()
    b_side_winrate = b_side_winrate.where(b_side_winrate["won"] == True).dropna()
    b_side_winrate = b_side_winrate.value_counts(subset = ["side"]).head()


    total = b_side_winrate[0] + r_side_winrate[0]
    red_victories = r_side_winrate[0]
    red_percentage = (red_victories*100/total)
    blue_victories = b_side_winrate[0]
    blue_percentage = (blue_victories*100/total)
    print(f"total games: {total} ")
    print(f"Blue side victories: {blue_victories} translates to {blue_percentage:.2f}% winrate")
    print(f"Red side victories: {red_victories} translates to {red_percentage:.2f}% winrate")
    
timed_blue_v_red_winrate()
```
